```python
from os.path import join
import pandas as pd
from constants import DATA_BASE, LDA_PATH
from utils import load, Unlemmatizer, TopicsLoader
```

### composing questionnair

```python
_min_ = 1
_max_ = 15

dpath = join(DATA_BASE, 'topics_questionnair')
with open(join(dpath, 'head.xml'), 'r') as fp:
    head = fp.read()
with open(join(dpath, 'tail.xml'), 'r') as fp:
    tail = fp.read()
with open(join(dpath, 'survey_layout.xml'), 'r') as fp:
    survey_layout = fp.read()
with open(join(dpath, 'introduction.xml'), 'r') as fp:
    introduction = fp.read()
with open(join(dpath, 'demographie.xml'), 'r') as fp:
    demographie = fp.read()

section_tail = '</section>\n'
question_tail = '  </question>\n'
page_tail = f'&lt;/page&gt;\n'

pieces = [head]

for s_i in range(10):
    section_t_head = (
        f'<section>\n'
        f'  <title>topics{s_i*10+1:02d}-{s_i*10+10}</title>\n' # 01-10
        f'  <attributes.specific program="ofb">\n'
        f'  <attr id="id">T{s_i}</attr>\n' # 0
        f'  <attr id="pos">{s_i+1}</attr>\n' # 1
        f'  </attributes.specific>\n'
    )
    pieces.append(section_t_head)
    for q_i in range(1, 11):
        topic_id = (10 * s_i) + q_i
        question_t_head = (
            f'  <question>\n'
            f'    <description>Topic {topic_id}</description>\n'
            f'    <class>checkbox</class>\n'
            f'    <title>Gruppe</title>\n'
            f'    <explanation />\n'
            f'    <inputPosition>right</inputPosition>\n'
            f'    <attributes.specific program="ofb">\n'
            f'    <attr id="id">{q_i}</attr>\n'
            f'    <attr id="pos">{q_i}</attr>\n'
            f'    <attr id="metrics.gap">1</attr>\n'
            f'    <attr id="restrictions.answered">yes</attr>\n'
            f'    <attr id="restrictions.max">{_max_}</attr>\n'
            f'    <attr id="restrictions.min">{_min_}</attr>\n'
            f'    <attr id="composition">rows</attr>\n'
            f'    <attr id="character">cards</attr>\n'
            f'    <attr id="columns">1</attr>\n'
            f'    <attr id="metrics.width_default">0</attr>\n'
            f'    </attributes.specific>\n'
        )
        pieces.append(question_t_head)
        for i_i in range(20):
            item = (
                f'    <item>\n'
                f'      <text>{topics.iloc[topic_id-1, i_i]}</text>\n'
                f'      <attributes.specific program="ofb">\n'
                f'      <attr id="id">{i_i+1}</attr>\n'
                f'      <attr id="pos">{i_i+1}</attr>\n'
                f'      </attributes.specific>\n'
                f'    </item>'
            )
            pieces.append(item)
        pieces.append(question_tail)
    pieces.append(section_tail)

for s_i in range(10):
    section_l_head = (
        f'<section>\n'
        f'  <title>labels{s_i*10+1:02d}-{s_i*10+10}</title>\n' # 01-10
        f'  <attributes.specific program="ofb">\n'
        f'  <attr id="id">L{s_i}</attr>\n' # 0
        f'  <attr id="pos">{s_i+1+10}</attr>\n' # 11
        f'  </attributes.specific>\n'
    )
    pieces.append(section_l_head)
    for q_i in range(1, 11):
        question_l = (
            f'  <question>\n'
            f'    <description>Label {10*s_i+q_i}</description>\n'
            f'    <class>text</class>\n'
            f'    <title />\n'
            f'    <explanation />\n'
            f'    <inputPosition>right</inputPosition>\n'
            f'    <attributes.specific program="ofb">\n'
            f'    <attr id="id">{q_i}</attr>\n'
            f'    <attr id="pos">{q_i}</attr>\n'
            f'    </attributes.specific>\n'
            f'    <item>\n'
            f'      <text>Vorschlag für Überschrift (optional)</text>\n'
            f'      <attributes.specific program="ofb">\n'
            f'      <attr id="id">1</attr>\n'
            f'      <attr id="pos">1</attr>\n'
            f'      </attributes.specific>\n'
            f'    </item>\n'
            f'  </question>\n'
        )
        pieces.append(question_l)
    pieces.append(section_tail)
    
pieces.append(demographie)
pieces.append(survey_layout)
pieces.append(introduction)

intID = 10
for p in range(10):
    pieces.append(f'&lt;!-- Seite {p+4} --&gt;')
    pieces.append(f'&lt;page intID=&quot;{intID}&quot;&gt;')
    intID += 1
    pieces.append(f'&lt;info&gt;{p*10+1}-{p*10+10}&lt;/info&gt;')
    pieces.append(f'&lt;html intID=&quot;{intID}&quot;&gt;&lt;![CDATA[')
    intID += 1
    pieces.append(
        f'&lt;p&gt;Markieren Sie die Begriffe, die Ihrer Meinung nach &lt;b&gt;NICHT&lt;/b&gt; in eine Gruppe gehören. '
        f'Markieren Sie mindestens {_min_} und maximal {_max_} Begriffe pro Gruppe als &lt;b&gt;unpassend&lt;/b&gt;.&lt;/p&gt;\n'
        f'&lt;hr&gt;\n'
        f']]&gt;&lt;/html&gt;\n'
    )
    for i in range(1, 11):
        pieces.append(f'&lt;question id=&quot;T{p}{i:02d}&quot; intID=&quot;{intID}&quot; /&gt;') # T001
        intID += 1
        pieces.append(f'&lt;question id=&quot;L{p}{i:02d}&quot; intID=&quot;{intID}&quot; /&gt;') # L001
        intID += 1
        pieces.append(f'&lt;html intID=&quot;{intID}&quot;&gt;&lt;![CDATA[')
        intID += 1
        pieces.append(
            f'&lt;hr&gt;\n'
            f']]&gt;&lt;/html&gt;'
        )
    pieces.append(page_tail)

pieces.append(tail)
joined = '\n'.join(pieces)
print(joined)

with open(join(dpath, 'survey_project.xml'), 'w') as fp:
    fp.write(joined)
```
