```python
ignore = ['good', 'categories', 'hashmap', 'links', 'meta', 'phrases', 'sample', 'cache']
re_ignore = r'.*?(' + '|'.join(ignore) + r').*?'
ignore_pattern = re.compile(re_ignore)

datasets = ['Europarl', 'FAZ', 'FOCUS', 'OnlineParticipation', 'PoliticalSpeeches', 'dewiki', 'dewac']
re_include = r'^(' + '|'.join(datasets) + r').*'
include_pattern = re.compile(re_include)
```

```python
# statistics for the ETL pipeline
stats = []
path = ETL_PATH
files = sorted([f for f in listdir(path)
                    if (
                        isfile(join(path, f))
                        and include_pattern.match(f)
                        and not ignore_pattern.search(f)
                       )
               ], key=lambda x: x.lower())

for name in files:
    gc.collect()
    full_path = join(path, name)
    if not isfile(full_path):
        continue

    print('reading', name)
    corpus = re.split(r'_.', name)[0]
    df = pd.read_pickle(join(path, name))
    stats.append((corpus, path, len(df), None, None, int(df.index[0]), df.title.iloc[0]))
    
with open(join(ETL_PATH, 'stats_etl.json'), 'w') as fp:
    json.dump(stats, fp, ensure_ascii=False)
```

```python
# adding combined and corpus sample for the ETL pipeline
# also replacing the corpus size for FAZ and FOCUS with the original crawls

# !!! run this only ONCE !!!

with open(join(ETL_PATH, 'stats_etl.json'), 'r') as fp:
    stats = json.load(fp)

path = ETL_PATH
dsets = ['dewac1', 'news', 'speeches']

for dataset in dsets:
    gc.collect()
    df = load(dataset)
    stats.append((dataset, path, len(df), None, None, int(df.index[0]), df.title.iloc[0]))
    
faz1 = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FAZ.pickle.gz'))
faz2 = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FAZ2.pickle.gz'))
faz_old = faz1.append(faz2)
focus_old = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FOCUS.pickle.gz'))
for stat in stats:
    if stat[0] == 'FAZ_combined.pickle':
        stat[0] == 'FAZ.pickle'
        print(len(faz_old))
        stat[2] = len(faz_old)
    elif stat[0] == 'FOCUS_cleansed.pickle':
        stat[0] == 'FOCUS.pickle'
        print(len(focus_old))
        stat[2] = len(focus_old)

with open(join(ETL_PATH, 'stats_etl.json'), 'w') as fp:
    json.dump(stats, fp, ensure_ascii=False)
```

```python
# statistics for the NLP pipeline
stats = []
path = NLP_PATH
files = sorted([f for f in listdir(path)
                    if (
                        isfile(join(path, f))
                        and include_pattern.match(f)
                        and not ignore_pattern.search(f)
                       )
               ], key=lambda x: x.lower())

for name in files:
    gc.collect()
    full_path = join(path, name)
    if not isfile(full_path):
        continue

    print('reading', name)
    corpus = re.split(r'_.', name)[0]
    df = pd.read_pickle(join(path, name))
    stats.append((corpus, path, len(np.unique(df.hash.values)), len(np.unique(df.sent_idx.values)), len(df), int(df.hash.iloc[0]), df.token.iloc[0]))
    gc.collect()
    
with open(join(ETL_PATH, 'stats_nlp.json'), 'w') as fp:
    json.dump(stats, fp, ensure_ascii=False)
```

```python
# adding combined and corpus sample for the NLP pipeline
# !!! run this only ONCE !!!

# sentence number for combined corpora is not correct!!!

with open(join(ETL_PATH, 'stats_nlp.json'), 'r') as fp:
    stats = json.load(fp)

path = NLP_PATH
dsets = ['dewac1', 'news', 'speeches']

for dataset in dsets:
    gc.collect()
    df = load(dataset, 'nlp')
    stats.append((dataset, path, len(np.unique(df.hash.values)), len(np.unique(df.sent_idx.values)), len(df), int(df.hash.iloc[0]), df.token.iloc[0]))
    gc.collect()

for stat in stats:
    stat[0].replace('_combined', '').replace('_cleansed', '')

with open(join(ETL_PATH, 'stats_nlp.json'), 'w') as fp:
    json.dump(stats, fp, ensure_ascii=False)
```

```python
# statistics for the phrase extraction pipeline
stats = []
path = SMPL_PATH
files = [f for f in listdir(path) 
         if include_pattern.match(f) and not ignore_pattern.search(f)
        ]

# include dewiki subdir
for name in files:
    full_path = join(path, name)
    if isdir(full_path):
        subdir = [join(name, f) for f in listdir(full_path)
                 if include_pattern.match(f) and not ignore_pattern.search(f)]
        files += subdir
        
for name in sorted(files, key=lambda x: x.lower()):
    gc.collect()
    full_path = join(path, name)
    if not isfile(full_path):
        continue

    print('reading', name)
    df = pd.read_pickle(full_path)
    corpus = re.split(r'_.', name)[0]
    doc_stats = (corpus, path, len(np.unique(df.hash.values)), len(np.unique(df.sent_idx.values)), len(df), int(df.hash.iloc[0]), df.token.iloc[0])
    print(doc_stats)
    stats.append(doc_stats)
    gc.collect()
    
with open(join(ETL_PATH, 'stats_smpl.json'), 'w') as fp:
    json.dump(stats, fp, ensure_ascii=False)
```

```python
# aggregating statistics
path = ETL_PATH
files = sorted([f for f in listdir(path) if re.match(r'stats.*?json$', f)])
stats = []
for name in files:
    full_path = join(path, name)
    with open(join(ETL_PATH, name), 'r') as fp:
        stats += json.load(fp)

df = pd.DataFrame.from_records(stats, columns=['file', 'path', 'nb_docs', 'nb_sents', 'nb_words', 'doc_hash', 'first_token'])
df['corpus'] = df.file.map(lambda x: re.split(r'[_./]', x)[0])
dfx = df.groupby(['corpus', 'path']).sum().drop('doc_hash', axis=1)
dfx[['nb_docs', 'nb_sents', 'nb_words']] = dfx[['nb_docs', 'nb_sents', 'nb_words']].astype('int64')
dfx.index = dfx.index.set_levels([s.lower() for s in dfx.index.levels[0]], level='corpus')
dfx = dfx.sort_index()
dfx.to_csv(join(ETL_PATH, 'stats.csv'))
dfx
```

```python
faz1 = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FAZ.pickle.gz'))
faz2 = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FAZ2.pickle.gz'))
faz_old = faz1.append(faz2)
faz_old = faz_old.rename(columns={'subset': 'section', 'dataset':'crawled'})
faz_old_counts = faz_old.groupby('section').count().query('crawled > 2')

faz_new = load('FAZ')
faz_new = faz_new.rename(columns={'subset': 'section', 'dataset':'filtered'})
faz_new_counts = faz_new.groupby('section').count().query('filtered > 2')

faz = pd.concat([faz_old_counts.crawled, faz_new_counts.filtered], axis=1)
faz['dataset'] = 'faz'
faz = faz.set_index('dataset', append=True).reorder_levels(['dataset', 'section'])
faz
```

```python
focus_old = pd.read_pickle(join(ETL_PATH, 'deprecated', 'FOCUS.pickle.gz'))
focus_old = focus_old.rename(columns={'subset': 'section', 'dataset':'crawled'})
focus_old_counts = focus_old.groupby('section').count().query('crawled > 2')

focus_new = load('FOCUS')
focus_new = focus_new.rename(columns={'subset': 'section', 'dataset':'filtered'})
focus_new_counts = focus_new.groupby('section').count().query('filtered > 2')

focus = pd.concat([focus_old_counts.crawled, focus_new_counts.filtered], axis=1)
focus['dataset'] = 'focus'
focus = focus.set_index('dataset', append=True).reorder_levels(['dataset', 'section'])
focus
```

```python
news = pd.concat([faz, focus])
news.to_csv(join(ETL_PATH, 'stats_news_sections.csv'))
news
```
